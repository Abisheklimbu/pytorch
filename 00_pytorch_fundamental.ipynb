# Tensor operation
1. Addition
2. subtraction
3. Multiplication(Element wise)
4. Division
5. Matrix Multiplication

```python
tensor = torch.tensor([5, 10, 15])
print(tensor)
```

```python
# addition
print(tensor + 10)
print(torch.add(tensor, 10))
```

```python
# subtraction
print(tensor - 10)
print(torch.sub(tensor, 10))
```

```python
# multiplication element-wise
print(f"{tensor} '*' {tensor}")
print(tensor * tensor)
print(torch.mul(tensor, tensor))
```

```python
# Division
print(tensor/5)
print(torch.div(tensor, 5))
```

# Matrix Multiplication

```python
%%time
torch.matmul(tensor, tensor)
```

```python
# By using loop
# %%time
value = 0
for i in range(len(tensor)):
  value += tensor[i] * tensor[i]
print(value)
```

# Two Main Rules while performing matrix multiplication needs to be well notified: 
1. The inner dimension should match 
* (3,2) @ (3, 3) will won't work 
* (3,3) @ (3,2) will work 

2. The result will be of outer dimension  
* (2,3) @ (3,2) -> (2,2) #@ is also used in place of matmul 

```python
import torch
```

```python
torch.matmul((torch.rand(2,3)), (torch.rand(3,2)))
```

```python
torch.matmul(torch.rand(2,2), torch.rand(3,2))
```

```python
# You can use transpose to change the shape of tensor eg:
tensor_A = torch.tensor([[10, 20], 
                         [30, 40], 
                         [50, 60]])

tensor_B = torch.tensor([[5, 10], 
                         [15, 20], 
                         [25, 30]])

print(f"Shape of tensor_A{tensor_A.shape} and Shape of tensor_B {tensor_B.shape}")
```

```python
# if we try to multiply we will get an error 
tensor.matmul(tensor_A, tensor_B)
```

```python
# Use transpose (.T) and again multiply 
tensor_B_T = tensor_B.T
print(f"{tensor_A} @ \n{tensor_B_T}")
print(f"Shape of tensor_A: {tensor_A.shape} Old Shape of tensor_B{tensor_B.shape} \nand Shape after transpose {tensor_B_T.shape}")
print(f"tensor_A: {tensor_A.shape} @ transpose(tensor_B_T) {tensor_B_T.shape}")
print(f"\nOutput")
torch.mm(tensor_A, tensor_B_T)
```

## Minimum, Maximum, Mean, Sum (Tensor Aggregation)

```python
# create tensor from 1 to 199 with gap of 10
x = torch.arange(1,200,10)
print(f"Random Tensor X : {x}")
print(f"Shape of Tensor X: {x.shape}")
```

```python
# minimum tensor in x 
print(f"Minimum tensor in X : {torch.min(x)}")
```

```python
# maximum tensor in x 
print(f"Maximum tensor in X : {torch.max(x)}")
```

```python
# mean tensor in x 
print(f"Mean of tensor X : {torch.mean(x.type(torch.float32))}")
```

```python
# sum tensor in x 
print(f"Sum of tensor X : {torch.sum(x)}")
```

```python
# Finding postion of min_value 
x.argmin(), torch.argmin(x)
print(f"Minimum index is {torch.argmin(x)} and value is {x[torch.argmin(x)]}")
```

```python
# Finding postion of max_value 
x.argmax(), torch.argmax(x)
print(f"Maximum index is {torch.argmax(x)} and value is {x[torch.argmax(x)]}")
```
