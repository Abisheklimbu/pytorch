# Tensor operation
1. Addition
2. subtraction
3. Multiplication(Element wise)
4. Division
5. Matrix Multiplication

```python
tensor = torch.tensor([5, 10, 15])
print(tensor)
```

```python
# addition
print(tensor + 10)
print(torch.add(tensor, 10))
```

```python
# subtraction
print(tensor - 10)
print(torch.sub(tensor, 10))
```

```python
# multiplication element-wise
print(f"{tensor} '*' {tensor}")
print(tensor * tensor)
print(torch.mul(tensor, tensor))
```

```python
# Division
print(tensor/5)
print(torch.div(tensor, 5))
```

# Matrix Multiplication

```python
%%time
torch.matmul(tensor, tensor)
```

```python
# By using loop
%%time
value = 0
for i in range(len(tensor)):
  value += tensor[i] * tensor[i]
print(value)
```
