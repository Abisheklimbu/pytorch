```python
# 2. Create a random tensor with shape (7, 7).
import torch
random_tensor_A = torch.rand(7,7)
print(random_tensor_A)
```

```python
# Perform a matrix multiplication on the tensor from 2 with another random tensor with shape (1, 7) (hint: you may have to transpose the 
# second tensor).
random_tensor_B = torch.rand(1,7).T
print(random_tensor_B)

torch.matmul(random_tensor_A,random_tensor_B)
```

```python
# 4. Set the random seed to 0 and do exercises 2 & 3 over again.
RANDOM_SEED = 0 

torch.manual_seed(RANDOM_SEED)
random_tensor_x = torch.rand(size=(7,7))

torch.manual_seed(RANDOM_SEED)
random_tensor_y = torch.rand(1,7).T

torch.matmul(random_tensor_x, random_tensor_y )
```

```python
# 5. Speaking of random seeds, we saw how to set it with torch.manual_seed() but is there a GPU equivalent? (hint: you'll need to look 
# into the documentation for torch.cuda for this one). If there is, set the GPU random seed to 1234.
# ans: torch.cuda.initial_seed() Return the current random seed of the current GPU.
```

```python
# 6. Create two random tensors of shape (2, 3) and send them both to the GPU (you'll need access to a GPU for this). Set torch.manual_seed(1234) 
# when creating the tensors (this doesn't have to be the GPU random seed).
torch.cuda.is_available()

device = "cuda" if torch.cuda.is_available() else "cpu"

torch.manual_seed(1234)
random_tensor_a = torch.rand(2,3)
random_tensor_a = random_tensor_a.to(device)
print(random_tensor_a, random_tensor_a.device)

torch.manual_seed(1234)
random_tensor_b = torch.rand(2,3)
random_tensor_b = random_tensor_b.to(device)
print(random_tensor_b, random_tensor_b.device)
```

```python
# 7. Perform a matrix multiplication on the tensors you created in 6 (again, you may have to adjust the shapes of one of the tensors).
random_tensor_b_reshaped = random_tensor_b.reshape(3,2)
print(random_tensor_b_reshaped)
matrix_mul = torch.matmul(random_tensor_a, random_tensor_b_reshaped)
print(matrix_mul)
```

```python
# 8. Find the maximum and minimum values of the output of 7.
print(f"Maximum: {torch.max(matrix_mul)} and Minimum: {torch.min(matrix_mul)}")

# 9. Find the maximum and minimum index values of the output of 7.
print(f"Maximum Index: {torch.argmax(matrix_mul)} and Minimum Index: {torch.argmin(matrix_mul)}")
```

```python
# 10. Make a random tensor with shape (1, 1, 1, 10) and then create a new tensor with all the 1 dimensions removed to be left with a tensor 
# of shape (10). Set the seed to 7 when you create it and print out the first tensor and it's shape as well as the second tensor and it's shape.

torch.manual_seed(7)
random_tensor_c = torch.rand(1, 1, 1, 10)
print(random_tensor_c, random_tensor_c.shape)
print()

torch.manual_seed(7)
random_tensor_d = torch.squeeze(random_tensor_c)
print(random_tensor_d, random_tensor_d.shape)

random_tensor_c == random_tensor_d
```
